## Matplotlib

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
x = [5, 2, 7]
y = [2, 16, 4]
plt.plot(x,y)
plt.title('Info')
plt.ylabel('Y axis')
plt.xlabel('X axis')
plt.show()
```

```python
#Sample data for energy consumption over 6 months (MWb)
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'June']
energu_consumption = [1200, 1300, 1100, 1500, 1400, 1600]
#Create a line plot
plt.plot(months, energu_consumption, marker='o', color='b', linestyle=':')
#Add titles and labels
plt.title('Energy Consumption over 6 months')
plt.xlabel('Month')
plt.ylabel('Energy Consumption (MWh)')
plt.show()
```

```python
# Sample data for energy consumption by different renewable sources
energy_sources = ['Solar', 'Wind', 'Hydropower', 'Biomass']
energy_values = [1200, 3400, 2900, 2500]
# Create a bar chart
plt.bar(energy_sources, energy_values, color='grey')
# Add titles and Labels
plt.title('Energy Consumption by Renewable Energy Source')
plt.xlabel('Energy Source')
plt.ylabel('Energy Consumption (MWh)')
plt.show()
```

```python
#Pie chart to show energy consumption distribution 
plt.pie(energy_values, labels=energy_sources, autopct='%1.1f%%', colors=['gold', 'lightblue', 'lightgreen', 'coral'])
#Add a title
plt.title('Energy Consumtion Share by source')
plt.show()                                                  
```

```python
#Sample data for energy consumption & carbon emmision 
carbon_emissions = [200, 500, 450, 300] # in kg CO2
#Create a scatter plot 
plt.scatter(energy_values, carbon_emissions, color='red')
#Add titles and labels 
plt.title('Energy Consumtion vs Carbon Emission')
plt.xlabel('Energy Consumption (MWh)')
plt.ylabel('Carbon Emission (kg CO2)')
plt.show()
```

```python
#Create a customize bar chart 
plt.bar(energy_sources, energy_values, color='lightblue', edgecolor='black', linewidth=2)
#Add titles and lables 
plt.title('Customize Energy Consumption By Source', fontsize=14, fontweight='bold')
plt.xlabel('Energy Source', fontsize=12)
plt.ylabel('Energy Consumption (MWh)', fontsize=12 )
#Adding grid and customize style 
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(['Energy Consumption'], loc='upper right')
plt.show()
```

## Seaborn

```python
import seaborn as sns
sns.set(style="whitegrid")
tips = sns.load_dataset("tips")
sns.barplot(x="day", y="total_bill", data=tips)
```

```python
#Advance of advance Seaborn plot 
sns.set(style="whitegrid")
tips = sns.load_dataset("tips")
numeric_data = tips.select_dtypes(include='number')
sns.heatmap(numeric_data.corr(), annot=True, cmap="coolwarm")
```

```python
import matplotlib.pyplot as plt 
#Creating a Dataframe for pair.plot 
data = {
    "Solar": [1200, 1500, 1300],
    "wind": [3400, 3600, 3200],
    "Hydropower": [2900, 3100, 2800],
    "Biomass": [2500, 2700, 2400]
}
df = pd.DataFrame(data)
#Create a pair plot
sns.pairplot(df)
plt.show()
```

```python
#Create a heatmap to show the corellation matrix
correlation_matrix = df.corr()
#Plot heatmap 
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidth=0.5)
#Add title
plt.title('Correlation Between Energy Sources')
plt.show()
```

```python
#Create a box plot 
sns.boxplot(data=df)
#Add a title 
plt.title('Distribution of Energy Consumption by Source')
plt.show()
```

```python
#Create a violin plot 
sns.violinplot(data=df)
#Add a title 
plt.title('Violin Plot of Energy Consumption Distributions')
plt.show()
```

```python
#Sample data for Energy Consumption and Carbon Emission
energy_values = [100, 200, 300, 400, 500]
carbon_emission = [10, 20, 30, 40, 50]
# Create a DataFrame with energy consumption and carbon emissions
df_reg = pd.DataFrame({
    'Energy Consumption': energy_values,
    'Carbon Emission': carbon_emission
})
#Create a regression plot
sns.regplot(x='Energy Consumption', y='Carbon Emission', data=df_reg)
#Add title 
plt.title('Energy Consumption vs Carbon Emission')
plt.xlabel('Energy Consumption')
plt.ylabel(' Carbon Emission')
plt.show()
```

```python
# Define the months
months = ['January', 'February', 'March', 'April', 'May', 'June']
#Creating a dataset for different regions and energy consumption over months
df_facet = pd.DataFrame({
    'Month': months * 3,
    'Energy Consumption': [1200, 1300, 1100, 1500, 1400, 1600] * 3,
    'Region': ['North', 'South', 'East'] * 6,
})
#Create a FacetGrid plot
g=sns.FacetGrid(df_facet, col='Region', hue='Region', height=4, aspect=1)
g.map(sns.lineplot, 'Month', 'Energy Consumption')
#Add title
plt.subplots_adjust(top=0.85)
g.fig.suptitle('Energy Consumption Over Time by Region')
plt.show()
```
